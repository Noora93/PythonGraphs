#### "Hospital Inpatient Prevention Quality Indicators (PDI) for Pediatric Discharges by Patient County (SPARCS): Beginning 2009"
##### Data shown in the graph displays the relation between Diabetes and UTI in 2009-2012

```python

import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline

plt.style.use('seaborn-colorblind')

dataset = pd.read_csv("../PDI.csv")
#print(dataset)
```

```python
dataset2 = dataset [['Year', 'PDI', 'Observed','County', 'Expected']]
dataset2
```

```python
## http://pandas.pydata.org/pandas-docs/stable/reshaping.html

filtered = dataset2[((dataset2.PDI == 'Diabetes Short-term Complications') | (dataset2.PDI == 'Urinary Tract Infection')) & (dataset2.County=='STATEWIDE')]
```

```python
pivoted = filtered.pivot(index='Year', columns='PDI', values='Observed')
```

```python
pivoted
```

```python
pivoted.plot(kind = 'line')
```

```python
pivoted.plot(kind = 'bar')
```
